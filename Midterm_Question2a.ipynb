```python
historical_pr = "/content/drive/My Drive/CIVE5363_Midterm_Personal/pr_day_CanESM5_historical_r1i1p1f1_gn_18500101-20141231.nc"
historical_pr = xr.open_dataset(historical_pr, engine='netcdf4')

historical_tas = "/content/drive/My Drive/CIVE5363_Midterm_Personal/tas_day_CanESM5_historical_r1i1p1f1_gn_18500101-20141231.nc"
historical_tas = xr.open_dataset(historical_tas, engine='netcdf4')

ssp585_pr1 = "/content/drive/My Drive/CIVE5363_Midterm_Personal/pr_day_CanESM5_ssp585_r1i1p1f1_gn_20150101-21001231.nc"
ssp585_pr1 = xr.open_dataset(ssp585_pr1, engine='netcdf4')

ssp585_pr2 = "/content/drive/My Drive/CIVE5363_Midterm_Personal/pr_day_CanESM5_ssp585_r1i1p1f1_gn_21010101-21801231.nc"
ssp585_pr2 = xr.open_dataset(ssp585_pr2, engine='netcdf4')

ssp585_tas1 = "/content/drive/My Drive/CIVE5363_Midterm_Personal/tas_day_CanESM5_ssp585_r1i1p1f1_gn_20150101-21001231.nc"
ssp585_tas1 = xr.open_dataset(ssp585_tas1, engine='netcdf4')

ssp585_tas2 = "/content/drive/My Drive/CIVE5363_Midterm_Personal/tas_day_CanESM5_ssp585_r1i1p1f1_gn_21010101-21801231.nc"
ssp585_tas2 = xr.open_dataset(ssp585_tas2, engine='netcdf4')
```

```python
desired_lat = 35.8
desired_lon = -91

historical_pr_location = historical_pr.sel(lat=desired_lat, lon=desired_lon, method='nearest')
historical_tas_location = historical_tas.sel(lat=desired_lat, lon=desired_lon, method='nearest')
ssp585_pr1_location = ssp585_pr1.sel(lat=desired_lat, lon=desired_lon, method='nearest')
ssp585_pr2_location = ssp585_pr2.sel(lat=desired_lat, lon=desired_lon, method='nearest')
ssp585_tas1_location = ssp585_tas1.sel(lat=desired_lat, lon=desired_lon, method='nearest')
ssp585_tas2_location = ssp585_tas2.sel(lat=desired_lat, lon=desired_lon, method='nearest')
```
