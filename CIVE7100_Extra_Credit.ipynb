```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Load the data
file_path = 'combined_signal 2.csv'
data = pd.read_csv(file_path, header=None, names=['value'])

# Generate time values
time = np.arange(0, 3, 0.01)

# Function to fit the polynomial
def polynomial_func(x, a, b, c, d):
    return a * x**5 + b * x**3 + c * x + d

# Fit the polynomial
params_poly, _ = curve_fit(polynomial_func, time, data['value'])

# Generate polynomial signal
poly_signal = polynomial_func(time, *params_poly)

# Function to fit the sinusoidal signal
def sinusoidal_func(x, a, b):
    return a * np.sin(b * x)

# Fit the sinusoidal signal
params_sin, _ = curve_fit(sinusoidal_func, time, data['value'])

# Generate sinusoidal signal
sin_signal = sinusoidal_func(time, *params_sin)

# Fit the linear trend
params_lin = np.polyfit(time, data['value'], 1)

# Generate linear trend
lin_trend = np.polyval(params_lin, time)

# Decompose the original signal into its components
seasonal_component = lin_trend + poly_signal + sin_signal
gaussian_noise = data['value'] - seasonal_component

# Calculate mean and variance of Gaussian noise
mean_gaussian = np.mean(gaussian_noise)
variance_gaussian = np.var(gaussian_noise)

# Print the parameters
print("Polynomial Parameters (a, b, c, d):", params_poly)
print("Sinusoidal Parameters (a, b):", params_sin)
print("Linear Trend Parameters (slope, intercept):", params_lin)
print("Mean of Gaussian Noise:", mean_gaussian)
print("Variance of Gaussian Noise:", variance_gaussian)

# Plotting the components
plt.figure(figsize=(16, 8))

plt.plot(time, lin_trend, label='Linear Trend')
plt.plot(time, poly_signal, label='Polynomial Signal')
plt.plot(time, sin_signal, label='Sinusoidal Signal')
plt.plot(time, gaussian_noise, label='Gaussian Noise')

plt.legend()
plt.ylim(-100, 100)
plt.xlabel('Time')
plt.ylabel('Value')
plt.title('Signal Decomposition')
plt.show()
```
